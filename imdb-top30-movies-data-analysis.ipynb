```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read CSV file (with read_csv'). If you're using this, replace with your csv dataset file location
movies = pd.read_csv('C:/dataanalysis/IMDB_top30.csv')

# View all columns in CSV
pd.set_option('display.max_columns', 20)

# Get Top 5 rows of dataframe
movies.head()
```

```python
# describe provides descriptive statistics of a DataFrame. Generating summary of central tendency, dispersion 
# and shape of distribution of the DataFrame
movies.describe(include= 'all')
```

```python
# Divide 'gross' and budget column to convert $ value to thousands $

movies['box_office']= movies['box_office'].floordiv(1000)
movies['budget']= movies['budget'].floordiv(1000)
```

```python
# Create new column named profit by subtracting 'budget' column from 'box_office' column

movies.insert(11, column='Profit', value = movies['box_office']-movies['budget'])
```

```python
print(movies[['Profit']].to_string(index=False)) 
```

```python
# Sorting dataframe by 'profit' column as reference using 'sort_values' function 
movies.sort_values(by=['Profit'], ascending = False, inplace=True, ignore_index=True)

# Get top 10 profitable movies by using based indexing
top_10=movies.loc[0:10, 'name']
top_10
```

```python
# set theme for graphical layout

sns.set_style("whitegrid")

# plot profit vs budget 

plt.figure(figsize=[8,6])
h= sns.jointplot(x='budget', y='Profit', data= movies)
h.set_axis_labels('budget', 'Profit', fontsize=12)
plt.subplots_adjust(top=0.9)
plt.suptitle('Scatter plot of Budget and Profit', fontsize=15)
plt.show()
```

```python
# find movies with negative profit
negative_profit = movies[movies['Profit'] < 0]
negative_profit.reset_index(drop=True)

# Get 10 movies sorted that took the biggest losses
negative_profit.sort_values(by='Profit')[0:10]
```

```python
# Change scale of metacritic metascore to match IMDB criteria of out of 10 
movies['metascore'] = movies['metascore']/10
```

```python
# add column average rating and find average ratings 
# Mean calculated along specific axis of DataFrame

movies.insert(7, column='average_rating', value = movies.loc[:,['metascore', 'rating']].mean(axis=1))
```

```python
#print average rating column

print(movies[['average_rating']].to_string(index=False)) 
```

```python
# Sort in descending order of average rating 
movies.sort_values(by ='average_rating', ascending = False, inplace = True)
```

```python
# print movies with their average rating and name 
print(movies[['average_rating', 'name']].to_string(index=False)) 
```

```python
# Find movies in dataframe where metascore and rating are matched by less than 1
# Get top movies where rating and metascore are closely matched and select top 10 from dataframe

UniversalAcclaim = movies[abs(movies['metascore'] - movies['rating']) < 1]
UniversalAcclaim[(abs(movies['metascore'] - movies['rating']) < 1) & (movies['average_rating'] >= 9)].sort_index()[0:10]
```

```python
# Sort by average rating 
# Display list 

UniversalAcclaim.sort_values(by = 'average_rating', ascending=False, inplace=True)
UniversalAcclaim.reset_index(drop= True, inplace= True)
UniversalAcclaim[0:10]
```

```python
# lets try using average vs absolte difference 
movies['avg_rating'] = (movies['rating'] + movies['metascore']) / 2
movies['absolute_diff'] = (movies['rating'] - movies['metascore'])

plt.figure(figsize=(10,5))

plt.scatter(movies['avg_rating'],movies['absolute_diff'], alpha=0.5)

# aadding annotations for movies on graph
# for i, name in enumerate(movies['name']):
#     plt.annotate(name, (movies['average_rating'][i],movies['absolute_diff'][i]),
#                 fontsize=8, rotation=7, ha='right')
    
plt.title('Distribution of Absolute Difference')
plt.xlabel('Average Rating')
plt.ylabel('Absolute Diffrenece')
plt.show()
```

```python
# lets try a more temporal analysis and separate movies by decade
print(movies[['name', 'year','rating','metascore']].dtypes)
movies['year'] = pd.to_numeric(movies['year'], errors ='coerce')
movies['rating'] = pd.to_numeric(movies['rating'], errors ='coerce')
movies['metascore'] = pd.to_numeric(movies['metascore'], errors ='coerce')
movies.dropna(subset=['rating', 'metascore'], inplace=True)

# calcaulation to convert year of release to decade
movies['decade'] = (movies['year'] // 10) *10

```

```python
print(movies[['decade', 'metascore', 'rating', 'name']].to_string(index=False)) 
```

```python
# calculate average and absolute diff
movies['avg'] = (movies['rating'] + movies['metascore']) / 2
movies['absolute_diff'] = (movies['rating'] - movies['metascore'])

decade_grouped = movies.groupby('decade').agg({'avg': 'mean', 'absolute_diff': 'mean', 'decade': 'count'})
fig, ax1 = plt.subplots()
plt.scatter(decade_grouped['avg'],decade_grouped['absolute_diff'], alpha=0.5)
for decade, row in decade_grouped.iterrows():
    plt.annotate("{} ({} movies)".format(decade, int(row['decade'])),
                 (row['avg'],
                  row['absolute_diff']),
                 fontsize=8, rotation=10, ha='right')

plt.title('Temporal Analysis')
plt.xlabel('Average Rating')
plt.ylabel('Absolute Diffrenece')

ax2 = ax1.twinx()
ax2.bar(decade_grouped.index,
        decade_grouped['decade'], alpha=0.3,
        color='gray', label ='Movie Count')
ax2.set_ylabel('Number of Movies',
color ='gray')

plt.show()
```

```python
# split genres using split as multiple genres are possible e.g. Action,Drama,Comedy etc. 
movies['genre'] = movies['genre'].str.split(',')

# count genre 
genre_counts = movies['genre'].explode().value_counts()

# print genre count
print(genre_counts)
```

```python
# plot genre distribution into bar graph 
genre_counts.plot(kind='bar', xlabel='Genre', ylabel='Count', title= 'Top 30 IMDB Movies: Genre Distribution') 
plt.show()
```
